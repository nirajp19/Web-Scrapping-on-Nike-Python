```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://www.nike.com/in/w/mens-jordan-shoes-37eefznik1zy7ok"

response = requests.get(url)
print(response)

soup = BeautifulSoup(response.text)
soup

shoe_elements = soup.find("div", class_="product-grid")

shoe_elements = soup.find_all('div', class_='product-card')
shoe_elements

names = []
prices = []

for shoe in shoe_elements:
    name = shoe.find('div', class_='product-card__title')
    price = shoe.find('div', class_='product-price')

    names.append(name)
    prices.append(price)

Shoes = {'Name': names,'Price': prices}

data = pd.DataFrame(Shoes)
data

csv_file = 'nike_jordan_shoes.csv'
data.to_csv(csv_file, index=False)

data
```
